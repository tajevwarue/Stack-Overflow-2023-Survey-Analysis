There are some columns allows users to input only one response. While there are other columns that allows users to give more than one response. To deal with these instances we'll create two functions ~ One that deals with only on response and the other that deals with multiple response. 

```python
# Single response instance 
data["MainBranch"].value_counts()
```

As shown above this column, each user is allowed to select only one option. 

```python
# Multiple response instance
data["CodingActivities"].value_counts()
```

As seen above, users were allowed to pick more than one options

```python
# Creating the one option function for percentage returns
def single_percent(column):
    # Getting value_counts and normalizing of the column
    col_values = column.value_counts(normalize = True)
    # Sorting the values in descending order
    col_values = col_values.sort_values(ascending = False)
    # Get percentage of values and round to 2 decimal places
    col_values = round(col_values * 100, 2)

    return col_values
```

```python
# Creating the one option function count returns
def single_count(column):
    # Getting value_counts and sorting the values in descending order 
    col_values = column.value_counts().sort_values(ascending = False)

    return col_values
```

```python
# Creating the multiple option function for percentage returns
def multiple_percentage(column):
    # Splitting and expanding the choices to columns
    cols_exp = column.str.split(";", expand = True)
    # Stack columns, take count, normalize values and sort values in descending order
    cols_stack = round(cols_exp.stack().value_counts(normalize = True).sort_values(ascending = False) * 100, 2)

    return cols_stack
```

```python
# Creating the multiple options function count returns
def multiple_count(column):
    # Split, expand, stack and value_counts and sorting the values in descending order
    cols_final = column.str.split(";", expand = True).stack().value_counts().sort_values(ascending = False)

    return cols_final
```

## MultiVariate Analysis

### 1. Differences and Similarity in using AI For WorkFlow

```python
# Function to extract Values from ai next columns
def ai(column):
    values = column.str.split(";", expand = True).stack().value_counts().sort_values(ascending = False)
    return values
```

```python
# Applying the function to the columns
very_different = ai(data["AINextVery different"])
neither_different = ai(data["AINextNeither different nor similar"])
somewhat_similar = ai(data["AINextSomewhat similar"])
very_similar = ai(data["AINextVery similar"])
somewhat_different = ai(data["AINextSomewhat different"])

# Combining the columns into one dataframe
ai_next = pd.concat([very_different, somewhat_different, neither_different, somewhat_similar, very_similar], axis = 1)
ai_next.columns = ["Very different", "Somewhat different", "Neither different nor similar",  "Somewhat similar",  "Very similar"]
ai_next
```

```python
# Visualizing the data
ai_next.plot(kind = "bar", stacked= True, figsize = (12, 6))
plt.title("Similarity/Differences in Development WorkFlow in 1 year from now using AI")
plt.xlabel("Task")
plt.ylabel("Frequency");
```

Most respondents anticipates difference in their workflow in 1 year from now as a result of AI tools
they are currently using

### 2. AI Tool using or intends to/not to use

```python
# Applying the function to the columns
interested_using = ai(data["AIToolInterested in Using"])
currently_using = ai(data["AIToolCurrently Using"])
not_interested_using = ai(data["AIToolNot interested in Using"])


# Combining the columns into one dataframe
ai_using = pd.concat([interested_using, currently_using, not_interested_using], axis = 1)
ai_using.columns = ["Interested in Using", "Currently Using", "Not Interested in Using"]
ai_using
```

```python
# Visualizing the data
ai_using.plot(kind = "bar", stacked= True, figsize = (12, 6))
plt.title("Areas using, intends to use or does not intend to use AI tools")
plt.xlabel("Task")
plt.ylabel("Frequency");
```

From the bar chart, most respondents are interested in using AI tools in all options except in Debugging and getting help (most respondents already currently using it), Collaborating with teammates (most respondents not interested in using it) and Writing code (most respondents currently using it).
